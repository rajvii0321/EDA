# Descriptive Statistics:

```python
import pandas as pd

df=pd.read_csv('ibm.csv')
df

mean=df.MonthlyIncome.mean()
median =df.MonthlyIncome.median()
std=df.MonthlyIncome.std()

print("Mean = ", mean)
print("Median = ", median)
print("Standard Deviation = ",std)
```

# Histograms and KDE:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df=pd.read_csv('ibm.csv')
df           
             
#HISTOGRAM

plt.hist(df["Age"], bins=(10,20,30,40,50,60), rwidth=0.95, label="AGE"),
plt.title('Histogram of Age of the Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.legend()
plt.show()

#KDE PLOT

sns.kdeplot(df["Age"], shade=True,bw_method=5)
plt.title('KDE Plot of Age of the Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()
```

# Correlation Heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

cars = sns.load_dataset('mpg')
cars

cars.groupby('origin').cylinders.value_counts()

origin_cyli = (cars
                 .groupby('origin')
                 .cylinders
                 .value_counts()
                 .unstack()
                 .fillna(0)
                )

sns.heatmap(origin_cyli, 
            cmap='PiYG',
            annot=True, 
            fmt=".0f",
            annot_kws={
                'fontsize':16,
                'fontweight':'bold',
                'fontfamily':'serif'
            }
           )

#For correlation 
#sns.heatmap(cars.corr())
```

# Multivariate Analysis:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('cars.csv')
df

x=df['mpg']
y=df['weight']

# using matplotlib
plt.scatter(x=x,y=y, label="MPG vs WEIGHT")
plt.xlabel("MPG")
plt.ylabel("WEIGHT")
plt.grid(1)
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

cars=sns.load_dataset('mpg')


x=cars['mpg']
y=cars['weight']
z=cars['origin']

sns.scatterplot(x=x,
                y=y,
                data=cars,
                hue=z,
                size=z,
                sizes=[10,50,100]
               )
plt.xlabel("MPG")
plt.ylabel("WEIGHT")
plt.title("MPG vs WEIGHT")
plt.grid(1)
plt.legend()
plt.show()
```
